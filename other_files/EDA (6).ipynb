```python
def fuzzy_matching(GT_doc, OCR_doc): # create files with matched sentences 
    # split GT and OCR-output of each row into sentences (based on full stops).
    print('started')
    gt_sentences = GT_doc.split(".")
    ocr_sentences = OCR_doc.split(".")
    print('gt len:', len(gt_sentences))
    print('ocr len:', len(ocr_sentences))
    
    match_list = []
    
    counter = 0
    while len(gt_sentences) != 0:
        j = 0
        counter += 1
        print('counter:', counter)
        while j < len(ocr_sentences):
            
            #if counter > 200:
            #    break
            try:
                #print('gt and ocr:', gt_sentences[0] , ocr_sentences[j])
                if SequenceMatcher(None, gt_sentences[0], ocr_sentences[j]).ratio() > 0.65:
                    #print('removed both:', gt_sentences[0], ocr_sentences[j])
                    if len(gt_sentences[0]) >= 4 and len(ocr_sentences[j]) >= 4:
                        match_list.extend([[gt_sentences[0], ocr_sentences[j]]])
                        gt_sentences.remove(gt_sentences[0])
                        ocr_sentences.remove(ocr_sentences[j])
                        #del gt_sentences[0]
                        #del ocr_sentences[j]
                    else:
                        #print('goes here')
                        gt_sentences.remove(gt_sentences[0])
                        ocr_sentences.remove(ocr_sentences[j])
                        #print('really removed?')
                        continue
                else:
                    j += 1
                    if j == len(ocr_sentences):
                        #print('removed:', gt_sentences[0])
                        gt_sentences.remove(gt_sentences[0])
                        #del gt_sentences[0]
            except Exception as e:
                print('except:', e)
                break

            
            
    
    matched_gt_sentences = []
    matched_ocr_sentences = []

    for sentence in match_list:
        if len(sentence[0]) or len(sentence[1]) > 0:
            matched_gt_sentences.append(sentence[0])
            matched_ocr_sentences.append(sentence[1])
            
    # .join() with lists
    
    print('returned')
    return matched_gt_sentences, matched_ocr_sentences

df.reset_index()

GT_matched_docs_datasets = []
OCR_matched_docs_datasets = []

for d in range(len(selections)):
    selections[d].reset_index()

    # match sentences GT and OCR
    GT_matched_docs_list = []
    OCR_matched_docs_list = []


    for i in range(len(selections[d])):
        matched_gt_sentences, matched_ocr_sentences = fuzzy_matching(df['gt text'][i][:100], df['ocr text'][i][:100])
        GT_matched_docs_list.append(matched_gt_sentences)
        OCR_matched_docs_list.append(matched_ocr_sentences) 

    GT_matched_docs_datasets.append(GT_matched_docs_list)
    OCR_matched_docs_datasets.append(OCR_matched_docs_list)
    print(len(GT_matched_docs_list))
        
#print(GT_matched_docs_list[0][:10])
#print(OCR_matched_docs_list[0][:10])
print(GT_matched_docs_datasets[0])
print(OCR_matched_docs_datasets[0])
GT_sentences_datasets = GT_matched_docs_datasets
OCR_sentences_datasets = OCR_matched_docs_datasets
```

```python
# word alignment
def word_alignment(GT, OCR):
    shortest_list = min([GT, OCR], key=len)
    longest_list = max([GT, OCR], key=len)
    min_ratio = 0.65

    for i in range(len(longest_list)):
        match = False
        #i = -i-1
        # check if there is a direct match:
        try:
            if OCR[i] == '' and GT[i] == '':
                del OCR[i]
                del GT[i]
            if SequenceMatcher(None, GT[i], OCR[i]).ratio() >= min_ratio:
                match = True
                continue
            elif match == False:
                # check if there is a match with a word before in OCR:
                try:
                    for j in range(1,10):
                        if i-j >= 0:
                            if j >= 5:
                                if (GT[i] >= 5) or (OCR[i-j] >= 5):
                                    pass
                                else:
                                    continue
                            if SequenceMatcher(None, GT[i], OCR[i-j]).ratio() >= min_ratio:
                                if not (GT[i] == '' and OCR[i-j] == ''):
                                    for k in range(0,j):
                                        OCR.insert(i-j, '')
                                    match = True
                                    continue
                        if i+j <= len(OCR):
                            if j >= 5:
                                if (GT[i] >= 5) or (OCR[i+j] >= 5):
                                    pass
                                else:
                                    continue
                            if SequenceMatcher(None, GT[i], OCR[i+j]).ratio() >= min_ratio:
                                if not (GT[i] == '' and OCR[i+j] == ''):
                                    for k in range(0,j):
                                        GT.insert(i, '')
                                    match = True
                                    continue
                except:
                    pass
        except IndexError:
            min([GT, OCR], key=len).append('')

        if match == False:
            try:
                if SequenceMatcher(None, GT[i], OCR[i]).ratio() >= 0.30:
                    pass
            except:
                pass
            else:
                print('bad match', GT[i], OCR[i])
                print('before forwards match GT:', GT)
                print('before forwards bad match OCR:', OCR)
                if i-1 >= 0:
                    GT.insert(i, '')
                else:
                    GT.insert(0, '')
                if i+1 <= len(GT):
                    OCR.insert(i+1, '')
                else:
                    GT.append('')
                print('after forwards match GT:', GT)
                print('after forwards bad match OCR:', OCR)
                
                continue
    
    print('after forwards GT:', GT)
    print('after forwards OCR:', OCR)
    
    while len(GT) != len(OCR):
        min([GT, OCR], key=len).append('')
    
    if True:
        
        shortest_list = min([GT, OCR], key=len)
        longest_list = max([GT, OCR], key=len)

        for i in range(len(longest_list)):
            i = -i-1
            print('i:', i)
            try:

                if (OCR[i] == '') ^ (GT[i] == ''):

                    continue
                if OCR[i] == ' ' and GT[i] == '':
                    del OCR[i]
                    del GT[i]
                if SequenceMatcher(None, GT[i], OCR[i]).ratio() > 0.65:
     
                    continue
                else:
                    # check if there is a match with a word before in OCR:
                    try:
                        for j in range(1,len(GT)):
                            if j >= 10:
                                extra_ratio = 0.1
                            else:
                                extra_ratio = 0
                            if i-j >= -len(OCR):
                                if j >= 5:
                                    if (len(GT[i]) >= 5) or (len(OCR[i-j]) >=5):
                                        pass
                                    else:
                                        continue
                            if GT[i] == '' or OCR[i-j] == '':
                                continue
                            if SequenceMatcher(None, GT[i], OCR[i-j]).ratio() > 0.65 + extra_ratio:
                                for k in range(0,j):
                                    if i+k < 0:
                                        GT.insert(i+1, '')
                                    else: 
                                        GT.append('')
                                break
                            if i+j <= -1:
                                if j >= 5:
                                    if (GT[i] >= 5) or (OCR[i+j] >=5):
                                        pass
                                    else:
                                        continue
                            if GT[i] == '' or OCR[i+j] == '':
                                continue
                            if SequenceMatcher(None, GT[i], OCR[i+j]).ratio() > 0.65 + extra_ratio:
                                for k in range(0,j):
                                    GT.insert(i, '')

                                break
                    except:
                        pass

            except IndexError:
                min([GT, OCR], key=len).append('')

            # if there is no match:
            if SequenceMatcher(None, GT[i], OCR[i]).ratio() > 0.30:
                pass
            else:
                print('bad match', GT[i], OCR[i])
                print('before backwards bad match GT:',  GT)
                print('before backwards bad match OCR:',  OCR)
                if (OCR[i] == '') or (GT[i] == ''):
                    continue
                else:
                    GT.insert(i, '')
                    OCR.insert(i+1, '')
                print('after backwards bad match GT:',  GT)
                print('after backwards bad match OCR:',  OCR)
                
                
    print('after backwards GT:',  GT)
    print('after backwards OCR:',  OCR)
                    
    while len(GT) != len(OCR):
        min([GT, OCR], key=len).append('')


    for i in range(len(GT)-1):
        try:
            print(i)
            print(GT[i], OCR[i])
            if (GT[i] == '') and (OCR[i] == ''):
                del GT[i]
                del OCR[i]
        except:
            break
    
    print('end GT:',  GT)
    print('end OCR:',  OCR)
    return GT, OCR

# word alignment
aligned_datasets_GT = []
aligned_datasets_OCR = []

for i in range(len(GT_sentences_datasets)): # for each dataset
    aligned_dataset_GT = []
    aligned_dataset_OCR = []
    for j in range(len(GT_sentences_datasets[i])): # for each document
        aligned_document_GT = []
        aligned_document_OCR = []
        for k in range(len(GT_sentences_datasets[i][j])): # for each sentence
            print('GT_sentences_datasets[i][j][k]:', GT_sentences_datasets[i][j][k])
            aligned_GT, aligned_OCR = word_alignment(GT_sentences_datasets[i][j][k].split(' '), OCR_sentences_datasets[i][j][k].split(' '))
            aligned_document_GT.append(aligned_GT)
            aligned_document_OCR.append(aligned_OCR)
        aligned_dataset_GT.append(aligned_document_GT)
        aligned_dataset_OCR.append(aligned_document_OCR)
    aligned_datasets_GT.append(aligned_dataset_GT)
    aligned_datasets_OCR.append(aligned_dataset_OCR)

print(len(aligned_datasets_GT))
print(len(aligned_datasets_GT[i]))
```

```python
#meta_df = pd.DataFrame(columns=['identifier', 'dataset', 'sentences GT', 'sentences OCR', 'percentage not matched GT'])
meta_df = pd.DataFrame(columns=['identifier', 'dataset'])

for i in range(len(selections)): # for each dataset
    selections[i].reset_index()
    for j in range(len(selections[i])): # for each document
        # identifier
        identifier = selections[i]['identifier'][j]
        # dataset
        dataset = i+1
        # count of sentences in GT >= 4 words
        sentences_gt = 0
        words_GT_sentences = []
        characters_GT_sentences = []
        for k in selections[i]['gt text'][j].split('.'):
            if len(k) >= 4:
                sentences_gt += 1
                words_GT_sentences.append(len(k))
                characters_GT_sentences.append(len(' '.join(k)))
        # count of sentences in OCR >= 4 words
        sentences_ocr = 0
        words_OCR_sentences = []
        characters_OCR_sentences = []
        for k in selections[i]['ocr text'][j].split('.'):
            if len(k) >= 4:
                sentences_ocr += 1
                words_OCR_sentences.append(len(k))
                characters_OCR_sentences.append(len(' '.join(k)))
        # count of matched sentences between GT and OCR
        matched_sentences = len(GT_matched_docs_datasets[i][j])
        # percentage of sentences not_matched compared to GT
        perc_missing = 1 - (matched_sentences / sentences_gt)
        
        # analysing the word alignment
        # word alignment for each sentence
        longest_streaks = []
        total_missing_words = []
        perc_missing_words = []
        for k in range(len(aligned_datasets_OCR[i][j])): # for each sentence
            streaks = []
            streak = 0
            for l in range(len(aligned_datasets_OCR[i][j][k])): # for each word
                if aligned_datasets_OCR[i][j][k][l] == '':
                    streak += 1
                else:
                    streaks.append(streak)
                    streak = 0
            longest_streaks.append(max(streaks))# longest streak of missing word in sentence
            total_missing_words.append(sum(streaks)) # total of missing words in a sentence
            perc_missing_words.append(sum(streaks) / len(aligned_datasets_OCR[i][j]))
        # statistics word alignment doc level:
        # find longest streak of all longest streaks:
        import statistics as s
        longest_streak = max(longest_streaks)
        avg_longest_streaks = s.mean(longest_streaks)
        avg_total_missing_words = s.mean(total_missing_words)
        avg_perc_missing_words = s.mean(perc_missing_words)
         
        
        meta_df = meta_df.append({'identifier': identifier, 'dataset': dataset, 'sentences GT': sentences_gt, 'average words sentences gt': s.mean(words_GT_sentences), 'total words GT': sum(words_GT_sentences), 'total characters GT': sum(characters_GT_sentences), \
                                  'sentences OCR': sentences_ocr, 'average words sentences ocr': s.mean(words_OCR_sentences), 'total words OCR': sum(words_OCR_sentences), 'total characters OCR': sum(characters_OCR_sentences), 'matched sentences': matched_sentences, '% not matched GT': perc_missing, \
                                 'longest streak missing words': longest_streak, 'average longest missing words streaks': avg_longest_streaks, 'average missing words': avg_total_missing_words, 'average % missing words': avg_perc_missing_words}, ignore_index = True)
            
            
        
print(meta_df)
print(meta_df[['sentences GT', 'matched sentences', '% not matched GT']])

# word alignment
aligned_datasets_GT = []
aligned_datasets_OCR = []
```
